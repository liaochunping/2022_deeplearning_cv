```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img, bigger=False):
    if bigger:
        plt.figure(figsize=(15,15)) 
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def sharpen(img, sigma=80):
    blur_img = cv2.GaussianBlur(img, (0, 0), sigma)
    usm = cv2.addWeighted(img, 1.5, blur_img, -0.5, 0)
    
    return usm
```

```python

def de_motionblur(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]], np.float32)
    dst = cv2.filter2D(image, -1, kernel=kernel)

    return dst
```

```python
def img_processing(img):
    img = sharpen(img)
    img = de_motionblur(img) 
    return img
```

```python
list = {"001.jpg","002.jpg","003.jpg","004.jpg","005.jpg","006.jpg","007.jpg","008.jpg"}
for i in list:
    origin_img = cv2.imread("./license_plate/{}".format(i))
    print("origin picture:")
    show_img(origin_img, bigger=False)

    result_img = img_processing(origin_img)
    print("result picture:")
    show_img(result_img, bigger=False)
    cv2.imwrite('{}'.format(i), result_img)
```
